```python
introndf=pd.read_csv('/storage/yhhuang/users/ruiyan/annotation/human_genomics_annotation/intron.bed',delimiter='\t',header=None)
introndf
```

```python
introndf['Feature']='intron'
introndf
```

```python
exondf=pd.read_csv('/storage/yhhuang/users/ruiyan/annotation/human_genomics_annotation/exon.bed',delimiter='\t',header=None)
exondf
```

```python
exondf['Feature']='exon'
exondf
```

```python
alldf=pd.concat([utr5df,utr3df,introndf,exondf],axis=0)
alldf
```

```python
alldf.columns=['chromosome','start','end','gene_id','score','strand','Feature']
alldf
```

```python
TSSdf=pd.read_csv('/storage/yhhuang/users/ruiyan/PBMC/scRNA-seq/scTSS_out/Unannotatioin_500_CD/count/afterfiltered.csv')
TSSdf
```

```python
TSSdf['TSS']=TSSdf['Unnamed: 0'].str.split('*',expand=True)[1].str.split('_',expand=True)[0]
TSSdf
```

```python
TSSdf.columns=['TSS_name','count','std','summit_count','unencoded_G','label','gene_id','summit_pos','TSS']
TSSdf
```

```python
TSSdf['TSS']=TSSdf['TSS'].astype('int64')
TSSdf['TSS']
```

```python
TSSdf=TSSdf[['TSS_name','gene_id','TSS']]
TSSdf
```

```python
## intron

featurels=[]
for i in range(0,len(TSSdf)):
    genename=TSSdf['gene_id'][i]
    selectrefdf=alldf[alldf['gene_id']==genename]
    selectrefdf.reset_index(inplace=True,drop=True)
    TSS=TSSdf['TSS'][i]
    name=TSSdf['TSS_name'][i]
    #print(name)
    
    for j in range(0,len(selectrefdf)):
        start=selectrefdf['start'][j]
        end=selectrefdf['end'][j]
        reffeature=selectrefdf['Feature'][j]
        if (TSS>start)&(TSS<end):
            featurels.append([name,reffeature])

            
```

```python
featurels

featureassigndf=pd.DataFrame(featurels)
featureassigndf
```

```python
featureassigndf.columns=['TSS_name','feature']
featureassigndf
```

```python
keepfirstdf=featureassigndf.drop_duplicates('TSS_name')
keepfirstdf
```

```python
keepfirstdf['feature'].value_counts()
```

```python
featureassigndf['feature'].value_counts()
```

# plot pie chart

```python
import pandas as pd
```

```python
df=pd.read_csv('/storage/yhhuang/users/ruiyan/PBMC/scRNA-seq/unannotation_TSS_addfeature_other_pythonpage.csv',index_col=0)
df
```

```python
countdf=pd.DataFrame(df['feature'].value_counts())
countdf
```

```python
countdf=countdf.reindex(['UTR5','intron','exon','UTR3'])
countdf
```

```python
countdf.index=["5'UTR","CDS-intron","CDS-exon","3'UTR"]
countdf
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.rcParams['legend.handlelength'] = 1
matplotlib.rcParams['legend.handleheight'] = 1
```

```python
fig, ax = plt.subplots(figsize=(6, 3))



#attentaion: ax.pie在countdf['feature'].values也会返回一些值给autopct,所以absolute_value需要一个val参数，这个val指的就是countdf['feature'].values
def absolute_value(val):
    a  = '{:.0f}'.format(val/100*countdf['feature'].values.sum())
    return a


ax.pie(countdf['feature'].values,colors=['teal','dimgray','coral','gold'],autopct=absolute_value)



ax.legend(countdf.index,
          loc="center left",
          bbox_to_anchor=(0.94, 0, 0.5, 1),frameon=False)
plt.title('Unannotation mode')

#fig.savefig("/storage/yhhuang/users/ruiyan/figure/PBMC_pct_unannotation.pdf", dpi=300, bbox_inches='tight')


plt.show()
```
